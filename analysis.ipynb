```julia
using CairoMakie
using JSON
using StatsBase
using SpecialFunctions
using LsqFit
```

```julia
function cycle_type(array::Vector)
    # Determines the cycle type of a given array.
    
    # Extract elements from the array
    arr_1 = array[1]
    arr_2 = array[2]
    arr_3 = array[3]
    
    # Check for different cycle types based on array elements
    if arr_1 != arr_2
        if arr_1 == arr_3
            return 2
        else
            return 0
        end
    else
        return 1
    end
end
```

```julia
function get_transition(folder_path, N)
    # Calculates transition statistics based on stored overlap data.
    
    # Initialize lists to store transition statistics
    alpha_list = []
    overlap_list = []
    percent_fixed = []
    percent_two_cycle = []
    std_list = []
    
    # Define minimum and maximum values for P
    P_min = Int(N * 0.1)
    P_max = Int(N * 0.2)

    # Iterate over P values
    for P in collect(P_min:1:P_max)
    
        # Construct file path
        file_path = folder_path * "overlap_$(N)_$(P).json"
        
        # Parse JSON file
        data = JSON.parsefile(file_path)
        mems = keys(data)

        overlaps = []

        P_fixed = 0
        P_two_cycle = 0

        # Iterate over memories
        for mem in mems
            
            overlaps_mem = data[mem]
            # Determine cycle type of the overlaps
            type_mem = cycle_type(overlaps_mem)

            # Update statistics based on cycle type
            if type_mem == 1
                P_fixed += 1
            elseif type_mem == 2
                P_two_cycle += 1
            end

            push!(overlaps, mean(overlaps_mem))

        end

        # Calculate transition statistics
        α = P / N
        mean_overlap = mean(overlaps)
        std_overlap = std(overlaps)

        # Store transition statistics in lists
        push!(alpha_list, α)
        push!(overlap_list, mean_overlap)
        push!(std_list, std_overlap)
        push!(percent_fixed, P_fixed / P)
        push!(percent_two_cycle, P_two_cycle / P)

    end
    
    # Return transition statistics
    return alpha_list, overlap_list, std_list, percent_fixed, percent_two_cycle
end
```

```julia
inverse_N_list = Float64[]
alpha_crit = Float64[]

f = Figure(resolution=(1200, 800))
ax1 = Axis(f[1, 1], xlabel=L"\alpha = \frac{P}{N}", ylabel="Overlap")
ax2 = Axis(f[1, 2], xlabel=L"\alpha = \frac{P}{N}", ylabel="Overlap Variance")
ax3 = Axis(f[2, 1], xlabel=L"\alpha = \frac{P}{N}", ylabel="Percentage Fixed Point (%)")
ax4 = Axis(f[2, 2], xlabel=L"\alpha = \frac{P}{N}", ylabel="Percentage Two-Cycle (%)")

for N in [500, 600, 700, 900, 1000, 2000]
    
    folder_path = "results/$(N)/"

    alpha_list, overlap_list, std_list, percent_fixed, percent_two_cycle = get_transition(folder_path, N)
    
    for (i, ov) in enumerate(overlap_list)
        if ov <= 0.75
            println(N, ",", alpha_list[i])
            push!(inverse_N_list, 1 / N)
            push!(alpha_crit, alpha_list[i])
            break
        end
    end

    scatter!(ax1, float.(alpha_list), float.(overlap_list), markersize=8, label="N = $(N)")
    scatter!(ax2, float.(alpha_list), float.(std_list), markersize=8, label="N = $(N)")
    scatter!(ax3, float.(alpha_list), 100 * float.(percent_fixed), markersize=7, label="N = $(N)")
    scatter!(ax4, float.(alpha_list), 100 * float.(percent_two_cycle), markersize=7, label="N = $(N)")
end

axislegend(ax1, position=:lb)
axislegend(ax2, position=:lt)
axislegend(ax3, position=:lb)
axislegend(ax4, position=:lt)
display(f)
```

```julia
function lineal(x, p)
    # Linear function with parameters p[1] and p[2]
    return p[1] .+ p[2] .* x
end

# Fit the linear function to the data
fit = curve_fit(lineal, inverse_N_list, alpha_crit, [0.14, 8])

# Generate points for the fitted line
x_fit = collect(0:0.0001:0.003)
y_fit = lineal(x_fit, fit.param)

# Display the fitted parameters
@show fit.param

# Create a new figure for the plot
f1 = Figure(resolution=(800, 400))
ax11 = Axis(f1[1, 1], limits=(0.0001, 0.003, 0.140, 0.170), xlabel=L"1/N", ylabel=L"\alpha_{crit}")

# Scatter plot of the data points
scatter!(ax11, inverse_N_list, alpha_crit)

# Plot the fitted line
lines!(ax11, x_fit, y_fit)

# Display the figure
display(f1)
```
